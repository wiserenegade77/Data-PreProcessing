```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X_full = pd.read_csv('./housing_train.csv', index_col='Id')
X_test_full = pd.read_csv('./housing_test.csv', index_col='Id')
```

```python
X_full.head()
```

```python
# Remove rows with missing target, seprate target from predictions
X_full.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X_full.SalePrice
X_full.drop(['SalePrice'], axis=1, inplace=True)

# To keep things Simple we will take only numerical predictors
X= X_full.select_dtypes(exclude=['object'])
X_test= X_test_full.select_dtypes(exclude=['object'])

# bresk off validation set from training data
X_train, X_valid, y_train, y_valid= train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
X_valid.head()
```

```python
# shape of training data 
print(X_train.shape)

# Number of missing values in each column of dataset
missing_val_count_by_column=(X_train.isnull().sum())
print(missing_val_count_by_column)

missing_val_count_by_column=(X_valid.isnull().sum())
print(missing_val_count_by_column)
```

```python
# first method of removing null values 
# remove the columns

X_train.drop(['GarageYrBlt'], axis=1, inplace=True)
X_train.drop(['MasVnrArea'], axis=1, inplace=True)
X_train.drop(['LotFrontage'], axis=1, inplace=True)

X_valid.drop(['GarageYrBlt'], axis=1, inplace=True)
X_valid.drop(['MasVnrArea'], axis=1, inplace=True)
X_valid.drop(['LotFrontage'], axis=1, inplace=True)

missing_val_count_by_column=(X_valid.isnull().sum())
print(missing_val_count_by_column)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train,X_valid,y_train,y_valid):
    model= RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train,y_train)
    pred= model.predict(X_valid)
    return mean_absolute_error(y_valid,pred)
    
```

```python
mae= score_dataset(X_train,X_valid,y_train,y_valid)
print(mae)
```

```python
# method 2 imputation

from sklearn.impute import SimpleImputer

my_imputer= SimpleImputer()
imputed_X_train= pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid= pd.DataFrame(my_imputer.transform(X_valid))


# Imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns


print(score_dataset(imputed_X_train,imputed_X_valid,y_train,y_valid))
```

```python
# extention of imputation 
# method 3
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()

# Make new columns indicating what will be imputed
cols_with_missing = [col for col in X_train.columns
                     if X_train[col].isnull().any()]
for col in cols_with_missing:
    X_train_plus[col + '_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col + '_was_missing'] = X_valid_plus[col].isnull()

# Imputation
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.transform(X_valid_plus))

# Imputation removed column names; put them back
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns

print("MAE from Approach 3 (An Extension to Imputation):")
print(score_dataset(imputed_X_train_plus, imputed_X_valid_plus, y_train, y_valid))
```
